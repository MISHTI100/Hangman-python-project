```python
import random
stages = [
    """
      -----
      |   |
      |
      |
      |
      |
   --------
    """,  # 6 lives left
    """
      -----
      |   |
      |   O
      |
      |
      |
   --------
    """,  # 5 lives left
    """
      -----
      |   |
      |   O
      |   |
      |
      |
   --------
    """,  # 4 lives left
    """
      -----
      |   |
      |   O
      |  /|
      |
      |
   --------
    """,  # 3 lives left
    """
      -----
      |   |
      |   O
      |  /|\\
      |
      |
   --------
    """,  # 2 lives left
    """
      -----
      |   |
      |   O
      |  /|\\
      |  /
      |
   --------
    """,  # 1 life left
    """
      -----
      |   |
      |   O
      |  /|\\
      |  / \\
      |
   --------
    """  # Game over
]
word_list = ["banana","potato","beautiful"]
lives = 6
chosen_word = random.choice(word_list)
print(chosen_word)
#emply list
display = []
for i in range (len(chosen_word)):
    display += '_'
print(display)
game_over = False
while not game_over:
    guessed_letter = input("Guess a letter:").lower()
    for position in range(len(chosen_word)):
        letter = chosen_word[position]
        if letter == guessed_letter:
           display[position] = guessed_letter
    print(display)
    if guessed_letter not in chosen_word:
        lives -=1
        if lives ==0:
            game_over = True
            print("You lose!")
    if '_' not in display:  #guessed all the correct letters
        game_over = True
        print('You win!')
    print(stages[lives])
```
